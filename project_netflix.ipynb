```python
colors = []

for lab, movie in netflix_movies.iterrows(): 
    genre = movie['genre']

    if genre == 'Children': 
        colors.append('red')
    elif genre == 'Documentaries': 
        colors.append('blue')
    elif genre == 'Stand-Up':
        colors.append('pink')
    else: colors.append('gray')
```

```python
# release_year = [movie['release_year'] for movie in netflix_movies]
# duration = [movie['duration'] for movie in netflix_movies]
release_year = []

for lab, movie in netflix_movies.iterrows(): 
    year = movie['release_year']
    release_year.append(year)
```

```python
duration = []

for lab, movie in netflix_movies.iterrows(): 
    dd = movie['duration']
    duration.append(dd)
```

```python
fig = plt.scatter(x = release_year, y = duration , c=colors)
fig = plt.title('Movie Duration by Year of Release')
fig = plt.xlabel('Release year')
fig = plt.ylabel('Duration (min)')
plt.show()
```
